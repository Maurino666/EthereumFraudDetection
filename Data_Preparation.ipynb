Procedo con la trasformazione delle variabili testuali in numeri e la rimozione delle colonne con varianza 0.

```python
from sklearn.preprocessing import LabelEncoder

# Rimuoviamo colonne inutili
df.drop(columns=['Unnamed: 0', 'Index', 'Address'], inplace=True, errors='ignore')

# Creiamo un oggetto LabelEncoder
label_encoder = LabelEncoder()

# Convertire variabili testuali in numeri solo se le colonne esistono
if "ERC20 most sent token type" in df.columns:
    df["ERC20_most_sent_token_type_encoded"] = label_encoder.fit_transform(df["ERC20 most sent token type"])
if "ERC20_most_rec_token_type" in df.columns:
    df["ERC20_most_rec_token_type_encoded"] = label_encoder.fit_transform(df["ERC20_most_rec_token_type"])

# Selezionare solo colonne numeriche per la correlazione
df_numeric = df.select_dtypes(include=['number'])

# Rimuovere colonne con varianza nulla (tutti i valori uguali)
df_numeric = df_numeric.loc[:, df_numeric.var() > 0]

# Calcolare la matrice di correlazione
correlation_matrix = df_numeric.corr()

# Heatmap della correlazione con FLAG
plt.figure(figsize=(8, 10))
sns.heatmap(
    correlation_matrix[["FLAG"]].sort_values(by="FLAG", ascending=False),
    annot=True, cmap="coolwarm", vmin=-1, vmax=1
)
plt.title("Correlazione tra le variabili e FLAG (senza colonne a varianza 0)")
plt.show()
```

## Studio delle variabili singole

Sembra necessario uno studio delle variabili singole in base alla FLAG.

Per visualizzare meglio le relazioni delle variabili numeriche, ho deciso di stampare le relative correlazioni di spearman e t-test, oltre che rappresentare le distribuzioni con boxplot e kde plot.

```python
import numpy as np
from scipy.stats import spearmanr, ttest_ind

# Identificare le colonne numeriche e categoriche
numerical_columns = df.select_dtypes(include=['number']).columns.tolist()
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()

# Analisi delle variabili numeriche
for col in numerical_columns:
    if col != "FLAG":  # Evitiamo di calcolare la correlazione di FLAG con sé stesso
        print(f"\n **Analisi della variabile numerica: {col}**\n")

        # Correlazione di Spearman
        spearman_corr, spearman_p = spearmanr(df[col], df["FLAG"], nan_policy='omit')
        print(f"📌 Spearman Correlation: {spearman_corr:.4f}, p-value: {spearman_p:.4f}")

        # Test T-test per confrontare le medie tra FLAG=0 e FLAG=1
        group_0 = df[df["FLAG"] == 0][col]
        group_1 = df[df["FLAG"] == 1][col]

        if len(group_0) > 1 and len(group_1) > 1:  # Verifica che ci siano abbastanza dati
            t_stat, t_p = ttest_ind(group_0, group_1, equal_var=False, nan_policy='omit')
            print(f"📌 T-test: t-statistic = {t_stat:.4f}, p-value = {t_p:.4f}")
        else:
            print("Non abbastanza dati per il T-test.")

        # Boxplot a scala symlog
        plt.figure(figsize=(8, 6))
        ax = sns.boxplot(x="FLAG", y=col, data=df)
        vals = df[col].dropna()
        val_min, val_max = vals.min(), vals.max()
        ax.set_yscale("symlog", linthresh=1)
        ax.set_ylim(val_min, val_max)

        plt.title(f"Distribuzione di {col} rispetto a FLAG")
        plt.show()

        group_0_log = np.log1p(group_0)  # log(1 + x)
        group_1_log = np.log1p(group_1)

        # kde plot con metodo silverman
        plt.figure(figsize=(8, 6))
        sns.kdeplot(group_0_log, label="FLAG = 0", fill=True, alpha=0.5, bw_method='silverman', clip=(0, None))
        sns.kdeplot(group_1_log, label="FLAG = 1", fill=True, alpha=0.5, bw_method='silverman', clip=(0, None))

        plt.legend()
        plt.title("KDE Plot su scala log(1 + x)")
        plt.xlabel("log(1 + Avg_min_between_sent_tnx)")
        plt.show()
```

Le analisi sopra mostano dei risultati interessanti, soprattutto per alcuni valori.
