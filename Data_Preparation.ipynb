
## Studio delle variabili singole

La variabile dipendente è binaria quindi sembra necessario uno studio delle variabili singole in base a essa.
Riteniamo che questo sia il modo più veloce per vedere se esistono correlazioni ovvie.

Innanzitutto eliminiamo le colonne con varianza 0.

```python
df = df.select_dtypes(include=['number']).loc[:, df.var() != 0]
```


Per visualizzare meglio le relazioni delle variabili numeriche, abbiamo deciso di stampare le relative correlazioni di spearman e t-test, oltre che rappresentare le distribuzioni con boxplot e kde plot. I dati hanno distribuzioni con lunghe code quindi è stato necessario mostrarli in scala logaritmica.

```python
from project_utils import symlog_transform, symlog_inverse, generate_log_ticks
from scipy.stats import spearmanr, ttest_ind, gaussian_kde

# Identificare le colonne numeriche e categoriche
numerical_columns = df.select_dtypes(include=['number']).columns.tolist()
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()

# Analisi delle variabili numeriche
for col in numerical_columns:
    if col != "FLAG":  # Evitiamo di calcolare la correlazione di FLAG con sé stesso
        print(f"\n **Analisi della variabile numerica: {col}**\n")

        linthresh = 1

        # Correlazione di Spearman
        spearman_corr, spearman_p = spearmanr(df[col], df["FLAG"], nan_policy='omit')
        print(f"Spearman Correlation: {spearman_corr:.4f}, p-value: {spearman_p:.4f}")

        # Test T-test per confrontare le medie tra FLAG=0 e FLAG=1
        group_0 = df[df["FLAG"] == 0][col]
        group_1 = df[df["FLAG"] == 1][col]

        var_0 = group_0.var()
        var_1 = group_1.var()

        if len(group_0) > 1 and len(group_1) > 1 and var_0 > 0 and var_1 > 0:
            t_stat, t_p = ttest_ind(group_0, group_1, equal_var=False, nan_policy='omit')
            print(f"T-test: t-statistic = {t_stat:.4f}, p-value = {t_p:.4f}")
        else:
            print(f"Varianza nulla o non abbastanza dati per il T-test su {col}.")

        if var_0 > 0 or var_1 > 0:
            # Boxplot a scala symlog
            plt.figure(figsize=(8, 6))
            ax = sns.boxplot(x="FLAG", y=col, data=df)
            vals = df[col].dropna()
            val_min, val_max = vals.min(), vals.max()
            ax.set_yscale("symlog", linthresh=linthresh)
            if val_min == val_max:
                # Per non rendere il grafico non rappresentabile quando i valori sono uguali
                ax.set_ylim(val_min - 1, val_max + 1)
            else:
                ax.set_ylim(val_min, val_max)

            plt.title(f"Distribuzione di {col} rispetto a FLAG")
            plt.show()

            # Generiamo i ticks
            tick_values_original = generate_log_ticks(val_min, val_max)

            # Convertiamo i valori originali nella scala symlog
            tick_values_transformed = [symlog_transform(v, linthresh = linthresh) for v in tick_values_original]

            # KDE Gaussiana
            group_0_log = symlog_transform(group_0)
            group_1_log = symlog_transform(group_1)

            # KDE plot con metodo silverman
            plt.figure(figsize=(8, 6))

            if var_0 > 0:
                sns.kdeplot(symlog_transform(group_0), label="FLAG = 0", fill=True, alpha=0.5, bw_method='silverman', clip=(0, None) if val_min >= 0 else None)
            else:
                print(f"Varianza nulla per FLAG=0 su {col}: impossibile eseguire KDE plot.")

            if var_1 > 0:
                sns.kdeplot(symlog_transform(group_1), label="FLAG = 1", fill=True, alpha=0.5, bw_method='silverman', clip=(0, None) if val_min >= 0 else None)
            else:
                print(f"Varianza nulla per FLAG=1 su {col}: impossibile eseguire KDE plot.")

            # Impostiamo i tick tondi
            plt.xticks(tick_values_transformed, [str(int(tick)) for tick in tick_values_original])

            plt.legend()
            plt.title(f"KDE Plot su scala symlog - {col}")
            plt.xlabel(col)
            plt.show()
        else:
            print("La variabile ha varianza nulla.")
```

Le analisi sopra mostrano dei risultati interessanti, soprattutto per alcuni valori che hanno distribuzioni differenti in base alla flag. Altri ,invece, mostrano una varianza nulla per una delle due variabili e una distribuzione molto concentrata.

Vogliamo vedere se il fatto di avere i token type a 0 o None possa avere qualche significato.
Possiamo farlo stampando un semplice grafico a barre che mostra la presenza di quei valori per ogni classe.

```python
#  Filtra i valori dove "Sent" è la stringa "0"
count_sent = df[df["ERC20 most sent token type"] == "0"].groupby("FLAG")["ERC20 most sent token type"].count()
#  Filtra i valori dove "Sent" è la stringa "None"
count_sent_none = df[df["ERC20 most sent token type"] == "None"].groupby("FLAG")["ERC20 most sent token type"].count()

# Filtra i valori dove "Received" è la stringa "0"
count_rec = df[df["ERC20_most_rec_token_type"] == "0"].groupby("FLAG")["ERC20_most_rec_token_type"].count()
# Filtra i valori dove "Received" è la stringa "None"
count_rec_none = df[df["ERC20_most_rec_token_type"] == "None"].groupby("FLAG")["ERC20_most_rec_token_type"].count()

# Rinominiamo le serie per chiarezza
count_sent.rename("Sent_0", inplace=True)
count_sent_none.rename("Sent_None", inplace=True)
count_rec.rename("Rec_0", inplace=True)
count_rec_none.rename("Rec_None", inplace=True)

# Calcoliamo il numero totale di record per ciascun FLAG
total_counts = df["FLAG"].value_counts()

# Forziamo i conteggi a reindicarsi in base a total_counts, riempiendo con 0
count_sent = count_sent.reindex(total_counts.index, fill_value=0)
count_sent_none = count_sent_none.reindex(total_counts.index, fill_value=0)
count_rec = count_rec.reindex(total_counts.index, fill_value=0)
count_rec_none = count_rec_none.reindex(total_counts.index, fill_value=0)

# Calcoliamo il conteggio dei valori restanti (quelli che non sono '0' o 'None')
count_sent_other = total_counts - (count_sent + count_sent_none)
count_rec_other = total_counts - (count_rec + count_rec_none)

# Rinominiamo le colonne
count_sent_other.rename("Sent_Other", inplace=True)
count_rec_other.rename("Rec_Other", inplace=True)

summary_df = pd.concat(
    [count_sent, count_sent_none, count_sent_other],
    axis=1
)
# Aggiorniamo il dataframe con i nuovi conteggi
summary_df = pd.concat([summary_df, count_rec, count_rec_none, count_rec_other], axis=1)

# Imposta l’ordine desiderato delle colonne
col_order = [
    "Sent_0", "Sent_None", "Sent_Other",
    "Rec_0", "Rec_None", "Rec_Other"
]

# Tieni solo queste colonne (nell’ordine che preferisci)
summary_df = summary_df[col_order]

# Creiamo un nuovo grafico a barre
summary_df.plot(kind='bar', figsize=(10, 6), rot=0)

plt.title("Conteggio di '0', 'None' e altri valori nei token inviati e ricevuti per FLAG")
plt.xlabel("FLAG")
plt.ylabel("Conteggio")
plt.legend(title="Tipo di valore")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostriamo il grafico
plt.show()

print(summary_df)
```

Si spiega la correlazione: le istanze con flag a 0 hanno più della metà dei valori settati a '0', mentre quelli con flag a 1 non hanno questi valori. Quindi il modello apprende questo pattern, aumentando il bias.

Riteniamo che queste due colonne possano introdurre un elevato bias, quindi riteniamo sia necessario rimuoverle.

## Feature Engineering

Adesso proveremo a fare un po' di feature engineering.

Innanzitutto volevamo provare a creare delle nuove features combinando quelle esistenti.

- 'sent_unique_ratio', come la proporzione di transazioni ad account unici su transazioni totali mandate;
- 'received_unique_ratio', come la proporzione di transazioni ad account unici su transazioni totali ricevute;
- 'tx_per_min', la media di transazioni per minuto.

```python
df["sent_unique_ratio"] = df["Unique Sent To Addresses"] / (df["Sent tnx"] + 1e-9)
df["received_unique_ratio"] = df["Unique Received From Addresses"] / (df["Received Tnx"] + 1e-9)
df["tx_per_min"] = df["total transactions (including tnx to create contract"] / (df["Time Diff between first and last (Mins)"] + 1e-9)
```

Queste feature non si sono rivelate utili. Il modello è peggiorato leggermente, che indica l'introduzione di rumore. In realtà quest cose si possono vedere anche dalle distribuzioni, che sono simili tra i valori di flag.

Adesso vorremmo provare invece a rimuovere le feature che hanno alta correlazione tra di loro.

```python
import numpy as np

#Calcoliamo la matrice di correlazione
corr_matrix = df.corr().abs()

#Creiamo una matrice triangolare superiore per evitare duplicati
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

#Identifichiamo le colonne con una correlazione > 0.9
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.9)]

#Rimuoviamo le colonne altamente correlate
df = df.drop(columns=to_drop)

# Create a mask for the lower triangle
mask = np.tril(np.ones_like(corr_matrix, dtype=bool))

plt.figure(figsize=(12, 10))  # Increase figure size
sns.heatmap(
    corr_matrix,
    mask=mask,
    cmap='viridis',
    annot=True,
    fmt=".1f",  # Format annotations to one decimal place
    linewidths=0.5,
    cbar_kws={'shrink': 0.8},
    annot_kws={"size": 8}  # Reduce font size in the squares
)
plt.title("Correlation Matrix (Triangular)", fontsize=16)
plt.xticks(fontsize=12, rotation=90)
plt.yticks(fontsize=12, rotation=0)
plt.tight_layout()  # Avoid text cutoff
plt.show()
```

Anche questa operazione sembra far peggiorare leggermente le prestazioni. Abbiamo confrontato l'accuracy nel training e nel testing per vedere se il modello andasse in overfitting. Sembra che non si verifichi questo problema con o senza queste colonne. Deduciamo quindi che l'eliminazione di queste colonne possa eliminare dati importanti.
