## Analysis and preparation of the data

Import delle librerie

```python
import pandas as pd

# Carica il file CSV (sostituisci 'tuo_file.csv' con il nome del tuo file)
df = pd.read_csv('data/transaction_dataset.csv')

# Mostra le prime righe
df.head()
```

Date le informazioni sopra si possono notare alcuni problemi:
- La colonna 'Unnamed: 0' è senza nome e non è necessaria;
- La colonna 'Index' è duplicata e non è necessaria data l'indicizzazione di Pandas;
- Le colonne 'ERC-20' contengono valori nulli.

Detto questo, adesso vorrei visualizzare le correlazioni degli attributi con il valore Flag.
Inizio con una gestione dei valori nulli, per consentire la creazione della matrice di correlazione.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Rimuoviamo colonne inutili
df.drop(columns=['Unnamed: 0', 'Index', 'Address'], inplace=True, errors='ignore')

# Pulizia dei dati: rimuovere o riempire i valori mancanti
df.dropna(inplace=True)  # Rimuove le righe con valori mancanti
# df.fillna(0, inplace=True)  # Alternativa: sostituire i NaN con 0

# Selezioniamo solo le colonne numeriche
df_numeric = df.select_dtypes(include=['number'])

# Calcolo della correlazione solo sulle colonne numeriche
correlation_matrix = df_numeric.corr()

# Ordinare le variabili più correlate al FLAG
flag_corr = correlation_matrix["FLAG"].sort_values(ascending=False)

# Visualizzare solo le variabili con correlazione significativa
plt.figure(figsize=(8, 10))
sns.heatmap(
    correlation_matrix[["FLAG"]]
    .sort_values(
        by="FLAG",
        ascending=False
    ),
    annot=True,
    cmap="coolwarm",
    vmin=-1,
    vmax=1
)
plt.title("Correlazione tra le variabili e FLAG")
plt.show()
```
